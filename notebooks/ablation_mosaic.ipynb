```python
import pickle

import matplotlib.pyplot as plt
```

```python
# Load all ablation methods
ablation_methods = ["zero", "mean", "zero-r", "mean-r"]
results = {}

for method in ablation_methods:
    file_path = (
        f"/work/results/ablation_mosaic/chronos-t5-base/{method}_wasserstein_mosaic.pkl"
    )
    with open(file_path, "rb") as f:
        results[method] = pickle.load(f)

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

# Plot each ablation method
for i, method in enumerate(ablation_methods):
    im = axes[i].imshow(results[method])
    axes[i].set_title(f"{method} Ablation")
    axes[i].set_xlabel("Head Index")
    axes[i].set_ylabel("Layer Index")
    fig.colorbar(im, ax=axes[i], fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()
```
